##### Code for the geting all columns from csv in the list format  and save in the txt file

````python
import pandas as pd

# The name of the CSV file you saved
csv_filename = '/content/MOCK_DATA.csv'

# The name of the output file you want to create
output_filename = 'all_rows_as_list.txt'

try:
    # Read the data from your CSV file
    df = pd.read_csv(csv_filename)

    # --- NEW: Drop the 'WorkEmail' column ---
    # The 'if' statement checks if the column exists before trying to drop it.
    if 'WorkEmail' in df.columns:
        df.drop(columns=['WorkEmail'], inplace=True)
        print("Successfully dropped the 'WorkEmail' column.")

    # Convert all the remaining DataFrame rows into a list of lists
    all_rows = df.values.tolist()

    # Write the list to the output text file
    with open(output_filename, 'w', encoding='utf-8') as f:
        f.write("[\n")
        for i, row in enumerate(all_rows):
            # Write each row followed by a comma and a newline
            f.write(f"    {str(row)}")
            if i < len(all_rows) - 1:
                f.write(",\n")
            else:
                f.write("\n")
        f.write("]\n")

    print(f"Successfully converted the data and saved it to '{output_filename}'!")

except FileNotFoundError:
    print(f"Error: The file '{csv_filename}' was not found.")
    print("Please make sure the CSV file is in the same directory as this script.")
except Exception as e:
    print(f"An error occurred: {e}")
````

````python
import json
def get_sql_query(genai_client, prompt, user_query):

  # https://www.geeksforgeeks.org/python/formatted-string-literals-f-strings-python/
  contents = f"""
  {prompt}

  Here's the user query in english you need to work on:
  {user_query}
  """
  response = genai_client.models.generate_content(model='gemini-2.5-flash', contents=contents)
  # print(response)

  # Access the usage_metadata attribute
  usage_metadata = response.usage_metadata

  # Print the different token counts
  print(f"Input Token Count: {usage_metadata.prompt_token_count}")
  print(f"Thoughts Token Count: {response.usage_metadata.thoughts_token_count}")
  print(f"Output Token Count: {usage_metadata.candidates_token_count}")
  print(f"Total Token Count: {usage_metadata.total_token_count}")

  output = json.loads(response.text.replace('```json', '').replace('```', ''))

  return output
````

````python
import sqlite3
import pandas as pd

def execute_query(query, db_name='employees.db'):

    conn = None
    try:
        # Connect to the databas
        conn = sqlite3.connect(db_name)
        cursor = conn.cursor()

        # Execute the query
        print(f"\nExecuting query on '{db_name}':\n{query}")
        cursor.execute(query)

        # Fetch all results
        results = cursor.fetchall()

        # Get column names from the cursor description
        columns = [description[0] for description in cursor.description]

        # Format results as a dataframe for easier use
        results_as_dict = [dict(zip(columns, row)) for row in results]
        results_df = pd.DataFrame(results_as_dict)

        print("Query executed successfully.")
        return results_df

    except sqlite3.Error as e:
        print(f"Database error executing query: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
    finally:
        if conn:
            conn.close()
````

````python
def text2sql(genai_client, prompt, user_query):
  output = get_sql_query(genai_client, prompt, user_query)
  if output['status'] == 'success':
    results = execute_query(output['response'])
    return results
  return output
````
